```python
import urllib2 
import simplejson
import json
import sys
import pandas as pd
import numpy as np
import random
import vincent
from vincent import Axis, AxisProperties, PropertySet, ValueRef
from pandas.io.json import json_normalize
import re
import openpyxl
from openpyxl.cell import get_column_letter
import requests
from StringIO import StringIO
vincent.core.initialize_notebook()

def loadjson(apiurl):
    jsondataurl = apiurl
    
    req = urllib2.Request(jsondataurl)
    opener = urllib2.build_opener()
    f = opener.open(req)
    dataframe = simplejson.load(f)
    return dataframe

def datasetreader(handle):
    coder = {}
    apiroot = "http://dpe.sandbox.socialhistoryservices.org/api/datasets?handle=" + handle
    dataset = loadjson(apiroot)    
    return dataset

fromurl = ''
if fromurl:
    handle = "hdl:10622/MRPFJ0"
    datasets = datasetreader(handle)
    for item in datasets:
        maindata = item['data']
else:
    # Geocoder
    handle = "hdl:10622/MRPFJ0"
    url = "http://dpe.sandbox.socialhistoryservices.org/api/datasets?handle=Panel[%27" + handle + "]&format=csv"
    res = requests.get(url)
    csvio = StringIO(res.content)
    maindata = pd.read_csv(csvio, sep='\t', dtype='unicode')
```

```python
def adjustdataframe(df):
    cols = df.columns
    checkcount = 0
    for colname in cols:
        noname = re.match("Unnamed", colname)
        if noname:
            checkcount = checkcount + 1
        
    if checkcount > 5:
        tmpdatacolumns = df.ix[1]        
        datacolumns = []
        for varcol in tmpdatacolumns:
            pyear = re.search(r'(\d{4})', str(varcol))            
            if pyear:
                thiscol = str(pyear.group(0))
            else:
                thiscol = varcol
            datacolumns.append(thiscol)
        
        df.columns = datacolumns
    return df

def loaddataset(apiroot, handle):
    url = apiroot + "/api/datasets?handle=Panel[%27" + handle + "]&format=csv"
    res = requests.get(url)
    csvio = StringIO(res.content)
    dataframe = pd.read_csv(csvio, sep='\t', dtype='unicode')
    
    maincode = 'Code'
    df = adjustdataframe(dataframe)    
    df.index = df[maincode]    

    #if df['Webmapper numeric code']:
    #    df.index = df['Webmapper numeric code']
    #elif df['Code']:
    #    df.index = df['Code']
        
    return df
```

```python
def dataframe_filter(codes, maincode, clioframe):    
    result = {}
    navicolumns = {}
    c = clioframe.columns.tolist()
    dataID = 0
    lastdata = []
    ycodes = []
                
    for code in codes:
        datafilter = clioframe[clioframe[maincode] == code]
        data = datafilter
        #.transpose()
        c = datafilter.columns.tolist()
        values = datafilter.get_values()
        # Delete extra columns
        for item in data:
            try:
                #print item
                year = int(item)
            except:
                del data[item]
                navicolumns[item] = values
        if dataID == 0:
            data.reset_index(col_fill=[code])
            totaldf = data
            dataID = dataID + 1
            ycodes.append(int(code))
        else:        
            data.reset_index(col_fill=[code])
            lastdata = data
            totaldf = pd.merge(totaldf, data, how='outer')
            ycodes.append(int(code))
        
        result[code] = data
        
    empty = {}
    for year in lastdata:
        empty[year] = ''
    
    return (totaldf, navicolumns, result, ycodes, empty)
```

```python
def get_info(clioframe):
    header = clioframe.columns
    countryinfo = {}    
    yearscolumns = {}
    line = []
    geo = {}
    for col in header:            
        try:
            year = int(col)
            yearscolumns[year] = year
        except:
            countryinfo[col] = clioframe[col]
            line = clioframe[col].values
               
    vocab = {}
    for geoid in clioframe.index:        
        geoinfo = []                
        geo[geoid] = ''
        lex = {}
        for colname in sorted(countryinfo):            
            sgeoid = str(geoid)            
            item = countryinfo[colname][sgeoid]                                    
            lex[colname] = item
        vocab[sgeoid] = lex
        
    return (countryinfo, line, vocab, yearscolumns)
            
```

```python
def save_historical_dataset(fullpath, geocoder, countryinfo, yearscolumns, dataset):
    wb = openpyxl.Workbook()
    ws = wb.get_active_sheet()
    ws.title = "Data"

    names = dataset.index
    i = 0
    col_width = 256 * 20
    # Forming header    
    order = []
    c = ws.cell(row=0, column=i)
    c.value = names[0]
    c = ws.cell(row=1, column=i)
    c.value = names[1]
    
    for newcolumn in sorted(countryinfo):        
        c = ws.cell(row=2, column=i)
        c.value = newcolumn
        c.style.font.bold = True
        order.append(newcolumn)
        ws.column_dimensions[get_column_letter(i+1)].width = 15
        i = i + 1
    for year in sorted(yearscolumns):        
        c = ws.cell(row=2, column=i)
        c.value = str(year)
        c.style.font.bold = True        
        i = i + 1        
            
    i = 2
    for idc in geocoder:
        i = i + 1
        metadata = geocoder[idc]
        ccode = geocoder[idc]['ccode']
        j = 0
        for columnname in order:            
            c = ws.cell(row=i, column=j)
            c.value = metadata[columnname]
            j = j + 1        
        for year in yearscolumns:            
            c = ws.cell(row=i, column=j)
            try:
                x = dataset.ix[str(ccode)][str(year)]                
                if str(x) == 'nan':
                    c.value = ''
                else:                    
                    c.value = x                                
            except:
                c.value = ''
            j = j + 1
        
    wb.save(fullpath)
    
    return fullpath
```

```python
def match_geocoders(maincode, tmpgeocoder, geo):
    geocoder = {}
    for item in tmpgeocoder:
        cid = item['id']
        name = item['name']
        ctrinfo = item['year']
        g = re.search(r'(\d+)\s+\-\-\s+(\d+)', ctrinfo)    
        try:
            startyear = g.group(1)
            endyear = g.group(2)
        except:
            (startyear, endyear) = ('', '')
    
        geoitem = {}
        geoitem[maincode] = str(cid)
        geoitem['start date'] = int(startyear)
        geoitem['end date'] = int(endyear)
        geoitem['country'] = name
    
        # Filling metadata
        for newcolumn in countryinfo:
            try:            
                geoitem[newcolumn] = str(geo[str(cid)][newcolumn])
            except:            
                geoitem[newcolumn] = ''            
    
        if int(startyear) < 2012:
            geocoder[cid] = geoitem
    
        try:
            geoinfra = geo[cid]
        except:        
            #print str(item)        
            skip = 1
        
    return geocoder
```
