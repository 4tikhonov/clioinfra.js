```python
import re

def findpid(handle):
    (pid, revid, cliohandle) = ('', '', '')
    ids = re.search(r'hdl\:\d+\/(\w+)', handle, re.M|re.I)
    if ids:
        pid = ids.group(1)
        files = re.search(r'hdl\:\d+\/\w+\:(\d+)\:(\d+)', handle, re.M|re.I)
        if files:
            fileid = files.group(1)
            revid = files.group(2)
            cliohandle = pid + str(fileid) + '_' + str(revid)
    return (pid, revid, cliohandle)

```
